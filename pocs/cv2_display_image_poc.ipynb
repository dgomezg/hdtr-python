```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img1 = cv2.cvtColor(cv2.imread("../samples/Toledo/sources/0001-enf.jpg"), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread("../samples/Toledo/sources/0002-enf.jpg"), cv2.COLOR_BGR2RGB)

blended_img = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

plt.imshow(blended_img, cmap="gray")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img1 = cv2.cvtColor(cv2.imread("../samples/BurrandStreetBridge-Vancouver/20060308061500.jpg"), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread("../samples/BurrandStreetBridge-Vancouver/20060308062500.jpg"), cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(cv2.imread("../samples/BurrandStreetBridge-Vancouver/20060308064500.jpg"), cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(cv2.imread("../samples/BurrandStreetBridge-Vancouver/20060308070500.jpg"), cv2.COLOR_BGR2RGB)

height, width = img2.shape[:2]
print("Height: "+str(height))
print("Width: "+str(width))

crop_img4 = img4[0:height, 0:width//4]
crop_img3 = img3[0:height, width//4:2*(width//4)]
crop_img2 = img2[0:height, 2*(width//4):3*(width//4)]
crop_img1 = img1[0:height, 3*(width//4):width]

crop_img4_img3 = np.concatenate((crop_img4, crop_img3), axis=1)
crop_img4_img3_img2 = np.concatenate((crop_img4_img3, crop_img2), axis=1)
crop_img4_img3_img2_img1 = np.concatenate((crop_img4_img3_img2, crop_img1), axis=1)

plt.imshow(crop_img4_img3_img2_img1, cmap="gray")
plt.xticks([]), plt.yticks([])
plt.show()
```
